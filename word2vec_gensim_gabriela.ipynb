```python
import pandas as pd

data_frame = pd.read_csv('dataset_sentencas_processadas.csv') #, skiprows=1)

# sentenca_original = sentenças obtidas do trabalho de Gabriela
# sentenca_processada1 = remoção das anotações feitas por Gabriela
# sentenca_processada2 = sem pontuações
# sentenca_processada3_1 = sem acentos
# sentenca_processada3_2 = sem stopwords
# sentenca_processada3_3 = letras minúsculas
# sentenca_processada3_4 = apenas as raizes das palavras (stemmer)
```

```python
data_frame.head(1)
```

```python
sentencas = data_frame['sentenca_processada2'].values.astype('U')
```

```python
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
import gzip
import gensim # Word2Vec
```

```python
documents = list()

for sentenca in sentencas:
    print(sentenca)
    documents.append(list(gensim.utils.simple_preprocess (sentenca)))
    
logging.info ("Done reading data file")

documents
```

```python
model = gensim.models.Word2Vec (documents, size=200, window=3, min_count=2, workers=2)
model.train(documents,total_examples=len(documents),epochs=100)
```

```python
w2v = dict(zip(model.wv.index2word, model.wv.syn0))
```

```python
w1 = "presidente"
model.wv.most_similar (positive=w1)
```

```python
w1 = ["polícia"]
model.wv.most_similar (positive=w1,topn=6)
```
